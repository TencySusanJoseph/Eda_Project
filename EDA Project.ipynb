```python
#2 Which products has maximum sale?
df.groupby('Product.name')['Sales'].sum().nlargest(5)
import seaborn as sns
import matplotlib.pyplot as plt
max_sale=df.groupby('Product.name')['Sales'].sum().nlargest(5)
sns.barplot(x=max_sale.values,y=max_sale.index,palette='Paired')
plt.title('Top 10 Products',size=20,pad=20)
plt.show()
```

```python
#15 Which shipment mode has highest Sales and Profit?
import pandas as pd
import matplotlib.pyplot as plt
df_stackb = df[['Ship.mode','Sales','Profit']]
# taking a groupby on category and then ship mode...
df_stackb = df_stackb.groupby(['Ship.mode']).sum().reset_index()
plt.figure(figsize=[10,15])
plt.subplot(3,1,1)
plt.bar(x=df_stackb['Ship.mode'], height=df_stackb['Sales'], color='#F05454')
plt.bar(x=df_stackb['Ship.mode'], height=df_stackb['Profit'], bottom=df_stackb['Sales'], color='#30475E')
plt.title("Sales & Profit Across Ship Modes", size=20, pad=20)
plt.legend(['Sales','Profit'])
```

##### Based on the analysis of the sales data, it's evident that the standard class shipping mode has both the highest sales and profit figures compared to other shipment modes. This conclusion indicates that standard class shipping is not only popular among customers but also contributes significantly to the profitability of the business.

```python
#16 How many days it will take for the order to be shipped after it has been placed?"
diff=df['Ship.date']-df['Order.date']
days=diff.mean()
print(days)
```

##### It will take approximately 3 days to be shipped after it has been placed.A three-day shipping timeframe suggests a relatively efficient delivery process, which could contribute to the popularity and profitability of standard class shipping.

```python
#17 Which is the top selling sub category?
Most_sold=df.groupby('Sub.category')['Sales'].sum().sort_index(ascending=False)
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data_set=df
plt.figure(figsize=(10,8))
sns.pointplot(x=Most_sold.index, y=Most_sold.values)
plt.xticks(rotation=90)
plt.xlabel('Sub.category')
plt.ylabel('Sales')
plt.title("Top selling Sub Category",size=20,pad=20)
plt.show()
```

##### Phones are the top selling subcategory. Recognizing phones as the top-selling subcategory suggests that there is high demand for these products among customer base.

```python
#18 What is the state with the highest profit?
high_prof=df.groupby('State')['Profit'].sum().nlargest(10)
print(high_prof)
```

```python
#19 Total profit with respect to region
#(we can denote region across profit as rgn_pft)
rgn_pft=df.groupby('Region')['Profit'].sum().sort_values(ascending=False)
print(rgn_pft)
import seaborn as sns
import matplotlib.pyplot as plt
plt.xticks(rotation=90)
sns.barplot(x=rgn_pft.index,y=rgn_pft.values,palette="Set2")
plt.title("Profit over Region",size=20,pad=20)
plt.show()
```
